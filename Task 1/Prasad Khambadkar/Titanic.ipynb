```python
import numpy as np
```

```python
columns_to_operate_on = ['age', 'fare', 'sibsp', 'parch']
```

```python
def perform_numpy_operations(dataframe, columns):
    for column in columns:
        column_data = dataframe[column].to_numpy()
        print(f"Column: {column}")
        print(f"Data: {column_data}")

        # mean
        mean_value = np.mean(column_data)
        print(f"Mean: {mean_value}")

        # median
        median_value = np.median(column_data)
        print(f"Median: {median_value}")

        # standard deviation
        std_value = np.std(column_data)
        print(f"Standard Deviation: {std_value}")

        # sum
        sum_value = np.sum(column_data)
        print(f"Sum: {sum_value}")

        # maximum and minimum
        max_value = np.max(column_data)
        min_value = np.min(column_data)
        print(f"Maximum: {max_value}")
        print(f"Minimum: {min_value}")

        print("=" * 40)

perform_numpy_operations(df, columns_to_operate_on)
```

```python
import matplotlib.pyplot as plt
```

```python
class_counts = df['pclass'].value_counts().sort_index()
plt.figure(figsize=(8, 6))
plt.bar(class_counts.index, class_counts.values, color='green', alpha=0.7)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Class Distribution')
plt.xticks(class_counts.index, labels=['1st', '2nd', '3rd'])
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['fare'], alpha=0.5, c='red')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare vs. Age')
plt.grid(True)
plt.show()
```

```python
class_counts = df['pclass'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=['1st', '2nd', '3rd'], autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Passenger Class Distribution')
plt.show()
```

```python
survival_by_class = df.groupby(['pclass', 'survived']).size().unstack();
plt.figure(figsize=(8, 6));
survival_by_class.plot(kind='bar', stacked=True, colormap='Set2')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'], loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
for pclass in df['pclass'].unique():
    subset = df[df['pclass'] == pclass]
    plt.scatter(subset['age'], subset['fare'], alpha=0.5, label=f'Class {pclass}')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare vs. Age by Passenger Class')
plt.legend()
plt.grid(True)
plt.show()
```
